```python
#Solución 4 Diego
from scipy.optimize import bisect
# Definición de g(x)
def g(x):
    return 16*x**2 - 32*x**3 + 16*x**4

# Definición de G(x)
def G(x):
    return (16/3)*x**3 - 8*x**4 + (16/5)*x**5

# Función a la cual queremos encontrar su raíz para obtener G^{-1}(u)
def root_func(x, u):
    return G(x) - u

# Generar variables aleatorias que siguen la distribución g(x)
def generate_random(n):
    samples = []
    for i in range(n):
        r = np.random.uniform(0, 1)
        try:
            x_val = newton(root_func, 0.5, args=(r,))
        except RuntimeError:
            # Before using bisect, check the signs at the endpoints
            if np.sign(root_func(0, r)) != np.sign(root_func(1, r)):
                x_val = bisect(root_func, 0, 1, args=(r,))
            else:
                # If signs are not different, skip this iteration (or handle differently)
                continue
        samples.append(x_val)
    return samples

# Genera 10000 puntos aleatorios
samples = generate_random(1000)

# Dibuja el histograma de las muestras y compáralo con g(x)
plt.hist(samples, density=True, bins=50, label="Samples", alpha=0.7)
xx = np.linspace(0, 1, 400)
plt.plot(xx, g(xx), 'r-', label="$g(x)$")
plt.xlim(0, 1)
plt.ylim(0, 1.5)
plt.legend()
plt.show()
```

Para usar el método de aceptación y rechazo, necesitamos una función \(g(x)\) tal que \(M \cdot g(x) \geq f(x)\) para todo \(x\) en el dominio de \(f(x)\), donde \(M\) es una constante.

Dado que $f(x) = 20x(1-x)^3$, el dominio de \(f(x)\) es $0 \leq x \leq 1$. Una elección común para \(g(x)\) es una distribución uniforme en el mismo dominio, es decir, \(g(x) = 1\) para \(0 \leq x \leq 1\).

Para encontrar \(M\), necesitamos maximizar \(f(x)\) en su dominio y luego dividir por \(g(x)\). En este caso, podemos usar cálculo para encontrar el máximo de \(f(x)\) en el intervalo [0,1].

Una vez que tengamos \(M\), el algoritmo de aceptación y rechazo es el siguiente:

1. Generar \(X\) de \(g(x)\) (en este caso, una distribución uniforme entre 0 y 1).
2. Generar \(U\) de una distribución uniforme entre 0 y 1.
3. Si $U \leq \frac{f(X)}{M \cdot g(X)}$, aceptar $X$; de lo contrario, volver al paso 1.

###### 9. Generación de variables aleatorias Poisson

La distribución de Poisson se define por la función de probabilidad:

$ P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!} $

Para demostrar que:

$ P(k+1)=\frac{\lambda}{k+1}P(k) $

Vamos a comenzar con el lado derecho de la ecuación y simplificarlo usando la definición de $P(k,\lambda)$.

La expresión simplificada de $\frac{\lambda}{k+1}P(k)$ es:

$ \frac{\lambda^{1 + k}}{e^{\lambda} (k! + k k!)} $

Ahora, vamos a expresar \(P(k+1,\lambda)\) y compararla con la expresión anterior.

La función $ P(k+1,\lambda) $ es simplemente $ P(1 + k, \lambda) $. 

Para demostrar que:

$ P(k+1) = \frac{\lambda}{k+1}P(k) $

Necesitamos mostrar que:

$ P(1 + k, \lambda) = \frac{\lambda^{1 + k}}{e^{\lambda} (k! + k k!)} $

Vamos a simplificar $ P(1 + k), \lambda)$ usando la definición de $ P(k,\lambda$.

Mis disculpas, parece que cometí un error al intentar simplificar la expresión. Vamos a proceder de otra manera.

Dada la definición de $ P(k,\lambda) $:

$ P(k+1,\lambda) = \frac{e^{-\lambda} \lambda^{k+1}}{(k+1)!} $

Vamos a expresar $ (k+1)! $ en términos de $ k! $ y simplificar la expresión.

La expresión simplificada de $ P(k+1,\lambda) $ es:

$ \frac{\lambda^{1 + k}}{e^{\lambda} (1 + k)!} $

Comparando esta expresión con la expresión simplificada de $ \frac{\lambda}{k+1}P(k)$:

Podemos ver que ambas expresiones no son iguales. Sin embargo, cometí un error en la simplificación anterior. La expresión correcta para \( \frac{\lambda}{k+1}P(k) \) es:

Comparando nuevamente, vemos que ambas expresiones son iguales, lo que demuestra la relación:

$ P(k+1) = \frac{\lambda}{k+1}P(k) $
