```python
#Soluci√≥n 4 Diego
from scipy.optimize import bisect
# Definici√≥n de g(x)
def g(x):
    return 16*x**2 - 32*x**3 + 16*x**4

# Definici√≥n de G(x)
def G(x):
    return (16/3)*x**3 - 8*x**4 + (16/5)*x**5

# Funci√≥n a la cual queremos encontrar su ra√≠z para obtener G^{-1}(u)
def root_func(x, u):
    return G(x) - u

# Generar variables aleatorias que siguen la distribuci√≥n g(x)
def generate_random(n):
    samples = []
    for i in range(n):
        r = np.random.uniform(0, 1)
        try:
            x_val = newton(root_func, 0.5, args=(r,))
        except RuntimeError:
            # Before using bisect, check the signs at the endpoints
            if np.sign(root_func(0, r)) != np.sign(root_func(1, r)):
                x_val = bisect(root_func, 0, 1, args=(r,))
            else:
                # If signs are not different, skip this iteration (or handle differently)
                continue
        samples.append(x_val)
    return samples

# Genera 10000 puntos aleatorios
samples = generate_random(1000)

# Dibuja el histograma de las muestras y comp√°ralo con g(x)
plt.hist(samples, density=True, bins=50, label="Samples", alpha=0.7)
xx = np.linspace(0, 1, 400)
plt.plot(xx, g(xx), 'r-', label="$g(x)$")
plt.xlim(0, 1)
plt.ylim(0, 1.5)
plt.legend()
plt.show()
```

Para usar el m√©todo de aceptaci√≥n y rechazo, necesitamos una funci√≥n \(g(x)\) tal que \(M \cdot g(x) \geq f(x)\) para todo \(x\) en el dominio de \(f(x)\), donde \(M\) es una constante.

Dado que $f(x) = 20x(1-x)^3$, el dominio de \(f(x)\) es $0 \leq x \leq 1$. Una elecci√≥n com√∫n para \(g(x)\) es una distribuci√≥n uniforme en el mismo dominio, es decir, \(g(x) = 1\) para \(0 \leq x \leq 1\).

Para encontrar \(M\), necesitamos maximizar \(f(x)\) en su dominio y luego dividir por \(g(x)\). En este caso, podemos usar c√°lculo para encontrar el m√°ximo de \(f(x)\) en el intervalo [0,1].

Una vez que tengamos \(M\), el algoritmo de aceptaci√≥n y rechazo es el siguiente:

1. Generar \(X\) de \(g(x)\) (en este caso, una distribuci√≥n uniforme entre 0 y 1).
2. Generar \(U\) de una distribuci√≥n uniforme entre 0 y 1.
3. Si $U \leq \frac{f(X)}{M \cdot g(X)}$, aceptar $X$; de lo contrario, volver al paso 1.

###### 9. Generaci√≥n de variables aleatorias Poisson

La distribuci√≥n de Poisson se define por la funci√≥n de probabilidad:

$ P(k,\lambda)=\frac{e^{-\lambda}(\lambda)^k}{k!} $

Para demostrar que:

$ P(k+1)=\frac{\lambda}{k+1}P(k) $

Vamos a comenzar con el lado derecho de la ecuaci√≥n y simplificarlo usando la definici√≥n de $P(k,\lambda)$.

La expresi√≥n simplificada de $\frac{\lambda}{k+1}P(k)$ es:

$ \frac{\lambda^{1 + k}}{e^{\lambda} (k! + k k!)} $

Ahora, vamos a expresar \(P(k+1,\lambda)\) y compararla con la expresi√≥n anterior.

La funci√≥n $ P(k+1,\lambda) $ es simplemente $ P(1 + k, \lambda) $. 

Para demostrar que:

$ P(k+1) = \frac{\lambda}{k+1}P(k) $

Necesitamos mostrar que:

$ P(1 + k, \lambda) = \frac{\lambda^{1 + k}}{e^{\lambda} (k! + k k!)} $

Vamos a simplificar $ P(1 + k), \lambda)$ usando la definici√≥n de $ P(k,\lambda$.

Mis disculpas, parece que comet√≠ un error al intentar simplificar la expresi√≥n. Vamos a proceder de otra manera.

Dada la definici√≥n de $ P(k,\lambda) $:

$ P(k+1,\lambda) = \frac{e^{-\lambda} \lambda^{k+1}}{(k+1)!} $

Vamos a expresar $ (k+1)! $ en t√©rminos de $ k! $ y simplificar la expresi√≥n.

La expresi√≥n simplificada de $ P(k+1,\lambda) $ es:

$ \frac{\lambda^{1 + k}}{e^{\lambda} (1 + k)!} $

Comparando esta expresi√≥n con la expresi√≥n simplificada de $ \frac{\lambda}{k+1}P(k)$:

$ \frac{\lambda^{1 + k}}{e^{\lambda} (k! + k k!)} $

Podemos ver que ambas expresiones no son iguales. Sin embargo, comet√≠ un error en la simplificaci√≥n anterior. La expresi√≥n correcta para \( \frac{\lambda}{k+1}P(k) \) es:

$ \frac{\lambda^{1 + k}}{e^{\lambda} (k+1) k!} $

Comparando nuevamente, vemos que ambas expresiones son iguales, lo que demuestra la relaci√≥n:

$ P(k+1) = \frac{\lambda}{k+1}P(k) $
